```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# 1. Drop irrelevant columns
df_mymodel = df.drop(columns=['ID', 'Customer_ID', 'Name', 'SSN', 'Type_of_Loan'])
```

```python
# 2. Encode categorical features
cat_cols = ['Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour']
le = LabelEncoder()
for col in cat_cols:
    df_mymodel[col] = le.fit_transform(df_mymodel[col])
```

```python
# 3. Encode the target variable (Credit_Score)
df_mymodel['Credit_Score'] = le.fit_transform(df_mymodel['Credit_Score'])  # Good = 1, Poor = 2, Standard = 0 (example)
```

```python
# 4. Separate features and target
X = df_mymodel.drop('Credit_Score', axis=1)
y = df_mymodel['Credit_Score']
```

```python
# 5. Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 6. Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import warnings
warnings.filterwarnings("ignore")
```

```python
# 1. Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
# 2. Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
# 3. Evaluation Function
def evaluate_model(y_true, y_pred, model_name):
    print(f"ðŸ” Evaluation Report for {model_name}")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Classification Report:\n", classification_report(y_true, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("-" * 50)
```

```python
# 4. Evaluate both models
evaluate_model(y_test, y_pred_lr, "Logistic Regression")
evaluate_model(y_test, y_pred_rf, "Random Forest")
```

```python
from xgboost import XGBClassifier
```

```python
# Initialize and train XGBoost
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
xgb.fit(X_train, y_train)
```

```python
# Predict
y_pred_xgb = xgb.predict(X_test)
```

```python
# Evaluate
evaluate_model(y_test, y_pred_xgb, "XGBoost Classifier")
```

```python

def predict_credit_score(user_input_dict):
    """
    Predict the credit score category for a single customer.
    Input: user_input_dict â€” a dictionary of input values for one customer
    Output: Predicted credit score category: 'Good', 'Standard', or 'Poor'
    """
    import pandas as pd
    import numpy as np

    # 1. Convert input dict to DataFrame
    user_df = pd.DataFrame([user_input_dict])

    # 2. Encode categorical features
    for col in ['Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour']:
        user_df[col] = le.transform(user_df[col])

    # 3. Select only model columns
    features = X.columns  # Same feature order used during training
    user_df = user_df[features]

    # 4. Scale numeric features
    user_scaled = scaler.transform(user_df)

    # 5. Predict
    pred_class = xgb.predict(user_scaled)[0]

    # 6. Convert prediction back to original label
    label = le.inverse_transform([pred_class])[0]

    return label
```
